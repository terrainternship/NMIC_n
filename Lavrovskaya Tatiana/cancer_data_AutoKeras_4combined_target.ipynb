```python
df = pd.read_csv('drive/MyDrive/datasets/cancer_dataset_preprocessed.csv')
```

```python
# Проверка уникальных значений каждого столбца
for col in df.columns:
    print(f'Уникальные значения столбца {col}: {df[col].unique()}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, concatenate
from tensorflow.keras.optimizers import Adam
import autokeras as ak

df['combined_target'] = df['Ответ эксперта (Лактионов)'] + '_' + df['Уверенность'] + df['Комментарий'] + df['Альтернатива, если 50% - обязательно']

# Разделение данных на обучающий и тестовый наборы
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Определение признаков и целевой переменной
X_train = train_df.drop(['Ответ эксперта (Лактионов)', 'Уверенность', 'Комментарий', 'Альтернатива, если 50% - обязательно', 'combined_target'], axis=1)
y_train = train_df['combined_target']

# Определение признаков и целевой переменной
X_test = test_df.drop(['Ответ эксперта (Лактионов)', 'Уверенность', 'Комментарий', 'Альтернатива, если 50% - обязательно', 'combined_target'], axis=1)
y_test = test_df['combined_target']

# Преобразование категориальных признаков
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Кодирование целевой переменной
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Разделение на обучающую и валидационную выборки
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train_encoded, test_size=0.2, random_state=42)

# AutoKeras модель
model = ak.StructuredDataClassifier(max_trials=2)

# Обучение модели
model.fit(X_train, y_train, epochs=5, validation_data=(X_val, y_val))

# Оценка модели
results = model.evaluate(X_test, y_test_encoded)
print(f' Accuracy: {results}')
```

```python
# Сохранение модели
model.export_model().save('/content/drive/MyDrive/NMIC/autokeras_structured_data_model', save_format='tf')
```
