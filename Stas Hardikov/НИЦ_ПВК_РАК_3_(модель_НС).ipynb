```python
# Список порогов для классификации типа расы
race_thresholds = [0,
              {'Азиатская'   : 0,
               'Европейская' : 1,
               'Другая'      : 2
              }]

# Список порогов для классификации пола
gender_thresholds = [0,
             {'Женский' : 0,
              'Мужской' : 1
             }]

# Список порогов для классификации возраста
age_thresholds = [0,
             {'До 70' : 0,
              '> 70'  : 1
             }]

# Список порогов для классификации статуса курения
smoking_status_thresholds = [0,
              {'В настоящее время'            : 0,
               'Курение в прошлом (более 1 месяца до 1 года)' : 1,
               'Курение в прошлом (бросил более 1 месяца назад)' : 2
              }]

# Список порогов для классификации ECOG (Eastern Cooperative Oncology Group) статуса
ecog_thresholds = [0,
             {'0-1' : 0,
              '2'   : 1
             }]

# Список порогов для классификации симптомной опухоли
tumor_symptom_thresholds = [0,
             {'да'  : 0,
              'нет' : 1
             }]

# Список порогов для классификации статуса гена KRAS
kras_gene_thresholds = [0,
             {'да'  : 0,
              'нет' : 1
             }]

# Список порогов для классификации статуса гена p53
p53_gene_thresholds = [0,
             {'да'  : 0,
              'нет' : 1
             }]

# Список порогов для классификации статуса гена STK11
stk11_gene_thresholds = [0,
             {'да'  : 0,
              'нет' : 1
             }]

# Список порогов для классификации статуса гена KEAP1
keap1_gene_thresholds = [0,
             {'да'  : 0,
              'нет' : 1
             }]

# Список порогов для классификации срока от окончания химиотерапии (ХЛТ)
xlt_duration_thresholds = [0,
              {'более 61 дня'     : 0,
               'до 42 дней'       : 1,
               'от 43 до 60 дней' : 2
              }]

# Список порогов для классификации молекулярного статуса
molecular_status_thresholds = [0,
              {'ALK позитивный'   : 0,
               'EGFR ex19'        : 1,
               'EGFR ex21'        : 2,
               'EGFR редкий вариант' : 3,
               'не исследовались'  : 4,
               'нет мутаций'      : 5
              }]

# Список порогов для классификации статуса PD-L1 (Programmed Death-Ligand 1)
pd_l1_status_thresholds = [0,
              {'Не исследовался' : 0,
               'Более 1%'        : 1,
               'Менее 1%'        : 2
              }]

# Список порогов для предпочтения пациента по ответу на терапию
patient_preference_thresholds = [0,
                            {'Результативность лечения'        : 0,
                             'сохранение качества жизни'       : 1
                            }]

# Список порогов для ответа эксперта (Лактионов) по выбору лечения
expert_answer_thresholds = [0,
                       {'Алектиниб'   : 0,
                        'Дурвалумаб'  : 1,
                        'Наблюдение'  : 2,
                        'Осимертиниб' : 3
                       }]
```

```python
# Список классов и их порогов для классификации атрибутов
class_descriptions = [race_thresholds, gender_thresholds, age_thresholds, smoking_status_thresholds, ecog_thresholds, tumor_symptom_thresholds,
                      kras_gene_thresholds, p53_gene_thresholds, stk11_gene_thresholds, keap1_gene_thresholds, xlt_duration_thresholds, molecular_status_thresholds,
                      pd_l1_status_thresholds, patient_preference_thresholds, expert_answer_thresholds]

# Для каждой классификации в списке class_descriptions
for class_desc in class_descriptions:
    # Если классификация представлена в виде списка (list)
    if isinstance(class_desc[1], list):
        # Обновляем количество классов (class_desc[0]) на основе длины списка плюс один
        class_desc[0] = len(class_desc[1]) + 1
    else:  # Если классификация представлена в виде словаря (dict)
        # Обновляем количество классов (class_desc[0]) на основе максимального значения в словаре плюс один
        class_desc[0] = max(class_desc[1].values()) + 1
```

Сначала определяется список class_descriptions, который содержит описания классов (пороги) для разных атрибутов.

Затем код перебирает каждое описание класса в списке class_descriptions.

Для каждого описания класса проверяется тип данных во втором элементе этого описания (class_desc[1]).

Если тип данных представленного описания класса - список (list), то код обновляет количество классов (class_desc[0]) на основе длины этого списка плюс один. Это предполагает, что каждый элемент списка представляет отдельный класс.

Если тип данных представленного описания класса - словарь (dict), то код обновляет количество классов (class_desc[0]) на основе максимального значения в этом словаре плюс один. В этом случае, словарь вероятно связывает категории с числовыми значениями, и максимальное значение используется для определения общего количества классов.

Код обновляет количество классов для каждого атрибута в соответствии с их типом данных и сохраняет обновленные значения в описаниях классов (class_descriptions). Это может быть полезно для дальнейшего анализа данных или классификации, где требуется знание количества классов для каждого атрибута.

```python


def integer_to_one_hot_encoding(input_value, class_descriptor):
    # Определение количества классов
    num_classes = class_descriptor[0]

    # Поиск соответствующего интервала для входного значения
    for i in range(num_classes - 1):
        if input_value < class_descriptor[1][i]:
            assigned_class = i  # Интервал найден - присвоение класса
            break
    else:  # Внимание: for/else
        assigned_class = num_classes - 1  # Интервал не найден - последний класс

    # Возврат результата в виде one-hot encoding-вектора
    return utils.to_categorical(assigned_class, num_classes)
```

```python
def str_to_multi(arg, class_dict):
    # Определение размерности выходного вектора
    num_classes = class_dict[0]

    # Создание нулевого вектора
    result = np.zeros(num_classes)

    # Поиск значения в словаре и, если найдено,
    # выставление 1. на нужной позиции
    for value, cls in class_dict[1].items():
        if value in arg:
            result[cls] = 1.

    return result
```

```python
# Последовательная модель НС
from tensorflow.keras.models import Sequential

# Основные слои
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization

# Утилиты для to_categorical()
from tensorflow.keras import utils

# Алгоритмы оптимизации для обучения модели
from tensorflow.keras.optimizers import Adam

# Библиотека для работы с массивами
import numpy as np

# Отрисовка графиков
import matplotlib.pyplot as plt

# Разделение данных на выборки
from sklearn.model_selection import train_test_split

# Для загрузки датасета
from sklearn.datasets import load_wine

# Отрисовка изображений в ноутбуке, а не в консоли или файле
%matplotlib inline
```

```python
# Разделение данных на тренировочную и временную (holdout) выборки
X_train, X_holdout, y_train, y_holdout = train_test_split(x_data, y_data, test_size=0.3, random_state=42)

# Затем разделите временную выборку на тестовую и валидационную
X_test, X_val, y_test, y_val = train_test_split(X_holdout, y_holdout, test_size=0.5, random_state=42)

print("Размерность X_train:", X_train.shape)
print("Размерность y_train:", y_train.shape)
print("Размерность X_test:", X_test.shape)
print("Размерность y_test:", y_test.shape)
print("Размерность X_val:", X_val.shape)
print("Размерность y_val:", y_val.shape)

```

# Создание модели

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout

model = Sequential()

# Добавление входного слоя
model.add(Dense(32, input_dim=14))  # 14 - количество признаков (вашей входной размерности)
model.add(Activation('relu'))  # Функция активации ReLU

model.add(Dropout(0.2))  # Dropout для регуляризации

model.add(Dense(16))
model.add(Activation('relu'))

# Выходной слой для классификации
model.add(Dense(4))  # 4 класса (количество методик лечения)
model.add(Activation('softmax'))  # Функция активации softmax для многоклассовой классификации

# Компиляция модели
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python





# Обучение модели
history = model.fit(X_train, y_train, epochs=8, batch_size=64, validation_data=(X_val, y_val))
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Потери на тестовых данных: {test_loss}")
print(f"Точность на тестовых данных: {test_accuracy}")
```

```python
import matplotlib.pyplot as plt

# Получение истории обучения
loss = history.history['loss']
val_loss = history.history['val_loss']
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Количество эпох
epochs = range(1, len(loss) + 1)

# Построение графика потерь
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Построение графика точности
plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
