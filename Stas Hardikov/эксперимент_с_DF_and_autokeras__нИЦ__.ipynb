```python
import pandas as pd
from IPython.display import display
from keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import csv
import autokeras as ak
from sklearn.preprocessing import LabelEncoder
```

```python
# Выбор столбцов с индексами с 2 по 14 (15 не включительно) с использованием iloc
X = df_copy.iloc[:, 1:15]
display(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, shuffle=True, random_state = 42)
```

```python
# Проверка x_train
if isinstance(x_train, np.ndarray):
    print("x_train является массивом NumPy.")
else:
    print("x_train не является массивом NumPy.")

# Проверка x_test
if isinstance(x_test, np.ndarray):
    print("x_test является массивом NumPy.")
else:
    print("x_test не является массивом NumPy.")

# Проверка y_train
if isinstance(y_train, np.ndarray):
    print("y_train является массивом NumPy.")
else:
    print("y_train не является массивом NumPy.")

# Проверка y_test
if isinstance(y_test, np.ndarray):
    print("y_test является массивом NumPy.")
else:
    print("y_test не является массивом NumPy.")
```

```python
#Эксперимент



x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, shuffle=True, random_state=42)
# Инициализация и настройка StructuredDataRegressor
regr = ak.StructuredDataRegressor(
    overwrite=True,
    max_trials=10,
    loss='mean_squared_error',
    metrics=['mean_absolute_error'],
    project_name='test',
)

# Подгонка модели на данных
regr.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), epochs=5)

# Оценка модели на тестовых данных
mse = regr.evaluate(x_test, y_test)[1]
print(f'Mean Squared Error on test data: {mse}')
```
