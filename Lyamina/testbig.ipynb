```python
import pandas as pd
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

# Предполагаем, что mytable уже существует

# Список названий столбцов, которые нужно преобразовать
columns_to_encode = ['Раса', 'Пол', 'Возраст', 'Статус курения', 'ECOG',
                     'Есть опухолевая нагрузка? (симптомная опухоль)',
                     'Ко-мутации KRAS', 'Ко-мутации p53.', 'Ко-мутации STK11',
                     'Ко-мутации KEAP1', 'Срок от окончания ХЛТ',
                     'Молекулярный статус (только для неплоскоклеточного рака)',
                     'PD-L1 статус', 'Предпочтение пациента по ответу на терапию', 'Ответ эксперта (Лактионов)']

# Создаем новый DataFrame для One-Hot Encoding
mytable_encoded = mytable.copy()

# Преобразуем каждый столбец в векторы One-Hot Encoding и добавим их в DataFrame
for column_name in columns_to_encode:
    # Проверка, является ли столбец категориальным (строковыми значениями)
    if mytable[column_name].dtype == 'object':
        # Используем Label Encoding для преобразования строковых значений в числа
        label_encoder = LabelEncoder()
        mytable_encoded[column_name] = label_encoder.fit_transform(mytable[column_name])

    integer_labels = mytable_encoded[column_name].astype(int)
    one_hot_vectors = to_categorical(integer_labels)

    # Создадим DataFrame из векторов One-Hot Encoding
    one_hot_df = pd.DataFrame(one_hot_vectors, columns=[f"{column_name}_{i}" for i in range(one_hot_vectors.shape[1])])

    # Объединим созданный DataFrame с исходным DataFrame
    mytable_encoded = pd.concat([mytable_encoded, one_hot_df], axis=1)

# Выведем обновленный DataFrame
print(mytable_encoded)
```

```python
mytable_encoded_colnames = mytable_encoded.columns.tolist()
# print(f'названия колонок: {len(mytable_encoded_colnames)}: {mytable_encoded_colnames}')

lst_count = len(mytable_encoded_colnames)
for i in range(lst_count):
  i_elem = mytable_encoded_colnames[i]
  print(f"{i}->{lst_count}: '{i_elem}'")
```

```python
#один

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense

# Отделим признаки (features) от нескольких целевых переменных (target variables)
features = mytable_encoded.iloc[:, 0:51]  # столбцы с 0 по 50 включительно
targets = mytable_encoded.iloc[:, 51:55]  # столбцы с 37 по 62 включительно

# Разделим данные на обучающую и тестовую выборки
features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size=0.2, random_state=42)

# Нормализуем данные (стандартизация)
scaler = StandardScaler()
features_train_scaled = scaler.fit_transform(features_train)
features_test_scaled = scaler.transform(features_test)

# Создадим простую нейросеть
model = Sequential()
model.add(Dense(32, input_dim=features_train.shape[1], activation='relu'))
model.add(Dense(targets_train.shape[1], activation='sigmoid'))  # Указываем количество выходных нейронов равным количеству целевых переменных

# Скомпилируем модель
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Обучим модель
model.fit(features_train_scaled, targets_train, epochs=10, batch_size=32, validation_data=(features_test_scaled, targets_test))
```
